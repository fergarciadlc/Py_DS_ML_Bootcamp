### We will order the house by price and drop the 1% of the top of that order, which results in a new data set of the 99% of the houses ignoring the more expensive ones

```python
len(df)*0.01
```

```python
non_top_1_perc = df.sort_values('price', ascending=False).iloc[216:]
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot('long','lat', hue = 'price',data=non_top_1_perc)
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot('long','lat', hue = 'price',data=non_top_1_perc, edgecolor=None, alpha=0.2, palette='RdYlGn')
```

Note that the prices start from RED color and become YELLOW and GREEN with the more expensive.
The edges show that having a House near the watter costs more than having it inland.
